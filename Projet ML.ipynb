### Principe de la boucle suivante:
On boucle sur la liste des équipes présentes dans les données.
1.on crée un df avec uniquement les matchs de cette équipe.
2.a partir du 4ème match, on regarde les performances des 3 derniers matchs de l'équipe (subtilité entre les colonnes à domicile et extérieur).
3.On fait la moyenne ou l'addition pour chaque indicateur, et on l'attribue dans un nouveau df portant le nom de l'équipe à la date du match (pour avoir les perf des 3 derniers matchs).

```python
for x in range(0,len(ListeEquipe)):
    equipe=ListeEquipe[x]
    nomDF=equipe
    tempo=data.loc[(data['HomeTeam'] == equipe) |  (data['AwayTeam'] == equipe)]
    tempo['HomeTeam'].count()

    tempo['Date']= pd.to_datetime(tempo['Date'])
    tempo=tempo.sort_values(by = 'Date', ascending = True)

    perf = pd.DataFrame(columns=['Date','Team','FTG','HTG','S','ST','C','Y','R','FTR','HTR'])
    nbligne=tempo['HomeTeam'].count()
    for u in range(3,nbligne):
        if tempo.iloc[u-3,1]==equipe:
            FTG1=tempo.iloc[u-3,3]
            HTG1=tempo.iloc[u-3,6]
            S1=tempo.iloc[u-3,9]
            ST1=tempo.iloc[u-3,11]
            C1=tempo.iloc[u-3,13]
            Y1=tempo.iloc[u-3,15]
            R1=tempo.iloc[u-3,17]
            if tempo.iloc[u-3,5]=='H':
                FTR1=3
            elif tempo.iloc[u-3,5]=='D':
                FTR1=1
            else:
                FTR1=0

            if tempo.iloc[u-3,8]=='H':
                HTR1=3
            elif tempo.iloc[u-3,8]=='D':
                HTR1=1
            else:
                HTR1=0


        if tempo.iloc[u-3,2]==equipe:
            FTG1=tempo.iloc[u-3,4]
            HTG1=tempo.iloc[u-3,7]
            S1=tempo.iloc[u-3,10]
            ST1=tempo.iloc[u-3,12]
            C1=tempo.iloc[u-3,14]
            Y1=tempo.iloc[u-3,16]
            R1=tempo.iloc[u-3,18]
            if tempo.iloc[u-3,5]=='H':
                FTR1=0
            elif tempo.iloc[u-3,5]=='D':
                FTR1=1
            else:
                FTR1=3

            if tempo.iloc[u-3,8]=='H':
                HTR1=0
            elif tempo.iloc[u-3,8]=='D':
                HTR1=1
            else:
                HTR1=3

        if tempo.iloc[u-2,1]==equipe:
            FTG2=tempo.iloc[u-2,3]
            HTG2=tempo.iloc[u-2,6]
            S2=tempo.iloc[u-2,9]
            ST2=tempo.iloc[u-2,11]
            C2=tempo.iloc[u-2,13]
            Y2=tempo.iloc[u-2,15]
            R2=tempo.iloc[u-2,17]
            if tempo.iloc[u-2,5]=='H':
                FTR2=3
            elif tempo.iloc[u-2,5]=='D':
                FTR2=1
            else:
                FTR2=0

            if tempo.iloc[u-2,8]=='H':
                HTR2=3
            elif tempo.iloc[u-2,8]=='D':
                HTR2=1
            else:
                HTR2=0

        if tempo.iloc[u-2,2]==equipe:
            FTG2=tempo.iloc[u-2,4]
            HTG2=tempo.iloc[u-2,7]
            S2=tempo.iloc[u-2,10]
            ST2=tempo.iloc[u-2,12]
            C2=tempo.iloc[u-2,14]
            Y2=tempo.iloc[u-2,16]
            R2=tempo.iloc[u-2,18]
            if tempo.iloc[u-2,5]=='H':
                FTR2=0
            elif tempo.iloc[u-2,5]=='D':
                FTR2=1
            else:
                FTR2=3

            if tempo.iloc[u-2,8]=='H':
                HTR2=0
            elif tempo.iloc[u-2,8]=='D':
                HTR2=1
            else:
                HTR2=3

        if tempo.iloc[u-1,1]==equipe:
            FTG3=tempo.iloc[u-1,3]
            HTG3=tempo.iloc[u-1,6]
            S3=tempo.iloc[u-1,9]
            ST3=tempo.iloc[u-1,11]
            C3=tempo.iloc[u-1,13]
            Y3=tempo.iloc[u-1,15]
            R3=tempo.iloc[u-1,17]
            if tempo.iloc[u-1,5]=='H':
                FTR3=3
            elif tempo.iloc[u-1,5]=='D':
                FTR3=1
            else:
                FTR3=0

            if tempo.iloc[u-1,8]=='H':
                HTR3=3
            elif tempo.iloc[u-1,8]=='D':
                HTR3=1
            else:
                HTR3=0

        if tempo.iloc[u-1,2]==equipe:
            FTG3=tempo.iloc[u-1,4]
            HTG3=tempo.iloc[u-1,7]
            S3=tempo.iloc[u-1,10]
            ST3=tempo.iloc[u-1,12]
            C3=tempo.iloc[u-1,14]
            Y3=tempo.iloc[u-1,16]
            R3=tempo.iloc[u-1,18]
            if tempo.iloc[u-1,5]=='H':
                FTR3=0
            elif tempo.iloc[u-1,5]=='D':
                FTR3=1
            else:
                FTR3=3

            if tempo.iloc[u-1,8]=='H':
                HTR3=0
            elif tempo.iloc[u-1,8]=='D':
                HTR3=1
            else:
                HTR3=3

        FTG=(FTG1+FTG2+FTG3)/3
        HTG=(HTG1+HTG2+HTG3)/3
        S=(S1+S2+S3)/3
        ST=(ST1+ST2+ST3)/3
        C=(C1+C2+C3)/3
        Y=(Y1+Y2+Y3)/3
        R=(R1+R2+R3)/3
        FTR=FTR1+FTR2+FTR3
        HTR=HTR1+HTR2+HTR3
        Date=tempo.iloc[u,0]
        perf = perf.append({'Date': Date,'Team': equipe,'FTG':FTG,'HTG':HTG,'S':S,'ST':ST,'C':C,'Y':Y,'R':R,'FTR':FTR,'HTR':HTR}, ignore_index=True)

    globals()[nomDF] = perf
```

## Partie modélisation

# regression logistique

```python
from sklearn.model_selection import train_test_split
data=dataAtro.iloc[:,4:23]
target=dataAtro.iloc[:,[23]]
x_train, x_test, y_train, y_test = train_test_split(data,target , test_size=0.25, random_state=50)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression().fit(x_train, y_train.values.ravel())
pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix
cmtx = pd.DataFrame(
    confusion_matrix(y_test, pred, labels=["D", "N","V"]), 
    index=["true:D", "true:N","true:V"], 
    columns=["pred:D", "pred:N","pred:V"]
)
print(cmtx)
print(classification_report(y_test, pred))
accuracy_score(y_test, pred)
```

# random forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 1000, max_depth=5, random_state = 42)
rf.fit(x_train, y_train.values.ravel())
```

```python
pred = rf.predict(x_test)
cmtx = pd.DataFrame(
    confusion_matrix(y_test, pred, labels=["D", "N","V"]), 
    index=["true:D", "true:N","true:V"], 
    columns=["pred:D", "pred:N","pred:V"]
)
print(cmtx)
print(classification_report(y_test, pred))
accuracy_score(y_test, pred)
```

# reseau de neurones

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='lbfgs', hidden_layer_sizes=5,alpha=0.001, random_state=10)
clf.fit(x_train, y_train.values.ravel())
```

```python
pred = clf.predict(x_test)
cmtx = pd.DataFrame(
    confusion_matrix(y_test, pred, labels=["D", "N","V"]), 
    index=["true:D", "true:N","true:V"], 
    columns=["pred:D", "pred:N","pred:V"]
)
print(cmtx)
print(classification_report(y_test, pred))
accuracy_score(y_test, pred)
```
